```python
import pandas as pd
import unidecode
```

```python
df = pd.read_csv('G:\My Drive\IT LOR\Phương Pháp Phát Triển Phần Mềm Hướng Đối Tượng\data_final_new.csv')
df
```

```python
for index, row in df.iterrows():
    value = row['district']
    value_new = unidecode.unidecode(value).replace(" ", "").lower()
    df.loc[index, 'district'] = value_new
df
```

```python

for index, row in df.iterrows():
    value = row['location']
    value_new = unidecode.unidecode(value).replace(" ", "").replace("(", "").replace(")", "").lower()
    df.loc[index, 'location'] = value_new
df
```

```python
for index, row in df.iterrows():
    value = row['type']
    value_new = unidecode.unidecode(value).replace(" ", "").replace(",","").lower()
    df.loc[index, 'type'] = value_new
df
```

```python
for index, row in df.iterrows():
    value = row['locations']
    value_new = unidecode.unidecode(value).replace(" ", "").replace("(", "").replace(")", "").replace(",", "_").lower()
    df.loc[index, 'locations'] = value_new
df
```

```python
df = df.sample(frac=1).reset_index(drop=True)
df
```

```python
df.to_csv('G:\My Drive\IT LOR\Phương Pháp Phát Triển Phần Mềm Hướng Đối Tượng\data_final_new.csv')
```
